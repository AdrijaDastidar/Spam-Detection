# Spam Detection

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
!gdown 'https://drive.google.com/uc?id=1PWL9JWCTa6a2N6TffUObhcl6TuLcwgI4'
data = pd.read_csv('mail_data.csv')
```

```python
data
```

```python
# replace the null values with a null string
mail_data = data.where((pd.notnull(data)),'')

# printing the first 5 rows of the dataframe
mail_data.head()
```

```python
# label spam mail as 0;  ham mail as 1;

mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
# separating the data as texts and label

X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the data - Test and Train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(X_train)
```

```python
# Feature extraction
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# Convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine": SVC()
}
```

```python
for model_name, model in models.items():

    # Train model
    model.fit(X_train_features, Y_train)

    # Predictions
    train_predictions = model.predict(X_train_features)
    test_predictions = model.predict(X_test_features)

    # Calculate accuracy
    train_accuracy = accuracy_score(Y_train, train_predictions)
    test_accuracy = accuracy_score(Y_test, test_predictions)

    # Print the results
    print(f"{model_name} - Accuracy on Training Data: {train_accuracy:.4f}")
    print(f"{model_name} - Accuracy on Test Data: {test_accuracy:.4f}\n")
```

```python
# Example input for prediction
input_mail = [
    "Hello, Hary how are you?",
    "Reply to our sms for a video mobile 750 mins UNLIMITED TEXT free money Reply or call now 08000930705"
]
```

```python
input_data_features = feature_extraction.transform(input_mail)
```

```python
# Making predictions with each model
for model_name, model in models.items():
    prediction = model.predict(input_data_features)
    print(f"\n{model_name} Prediction:")
    for i, pred in enumerate(prediction):
        if pred == 1:
            print(f"Mail {i + 1}: Ham Mail")
        else:
            print(f"Mail {i + 1}: Spam Mail")
```
